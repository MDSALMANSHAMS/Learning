### Translation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('art.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
print("image shape:",image.shape)

# Define the translation matrix
tx, ty = -200, 100  # translation in x and y directions
translation_matrix = np.float32([[1, 0, tx], [0, 1, ty]])
print("translation_matrix:", translation_matrix)

# Apply the translation
translated_image = cv2.warpAffine(image, translation_matrix, (image.shape[1], image.shape[0]))

# Display the original and translated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(translated_image)
plt.title('Translated Image')
plt.axis('off')

plt.show()
```

### Scaling

```python
import cv2
import matplotlib.pyplot as plt

# Load an image
#image = cv2.imread('art.jpg')
image = cv2.imread(r'F:\My_projects\Learning\Python_imageProcessing\nature.jpg')

image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Define the scaling factors
scale_x, scale_y = 0.2, 0.2

# Get the new dimensions
new_width = int(image.shape[1] * scale_x)
new_height = int(image.shape[0] * scale_y)
new_dimensions = (new_width, new_height)

# Apply the scaling
scaled_image = cv2.resize(image, new_dimensions, interpolation=cv2.INTER_LINEAR)

# Display the original and scaled images
fig, axs = plt.subplots(1, 2, figsize=(15, 5))
axs[0].imshow(image)
axs[0].set_title('Original Image')
axs[0].axis('off')

axs[1].imshow(scaled_image)
axs[1].set_title('Scaled Image')
axs[1].axis('off')

plt.show()
```

```python
# Display the original and scaled images using cv2.imshow
cv2.imshow('Original Image', image)
cv2.imshow('Scaled Image', scaled_image)

# Wait for a key press and close the windows
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### Flip

```python
import cv2
import matplotlib.pyplot as plt

# Load an image
#image = cv2.imread('art.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Reflect across the x-axis
reflected_image_x = cv2.flip(image, 0)  # 0 means flipping around the x-axis

# Reflect across the y-axis
reflected_image_y = cv2.flip(image, 1)  # 1 means flipping around the y-axis

# Reflect across both axes
reflected_image_xy = cv2.flip(image, -1)  # -1 means flipping around both axes

# Display the original and reflected images
plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 4, 2)
plt.imshow(reflected_image_x)
plt.title('Reflected Image X')
plt.axis('off')

plt.subplot(1, 4, 3)
plt.imshow(reflected_image_y)
plt.title('Reflected Image Y')
plt.axis('off')

plt.subplot(1, 4, 4)
plt.imshow(reflected_image_xy)
plt.title('Reflected Image XY')
plt.axis('off')

plt.show()
```

### Rotation

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('art.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Define the center of rotation, angle of rotation, and scale
center = (image.shape[1] // 2, image.shape[0] // 2)
angle = 45  # degrees
scale = 1.0

# Get the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)

# Apply the rotation
rotated_image = cv2.warpAffine(image, rotation_matrix, (image.shape[1], image.shape[0]))

# Display the original and rotated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(rotated_image)
plt.title('Rotated Image')
plt.axis('off')

plt.show()
```

### Shearing

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('art.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Define the shearing matrix
shear_factor = 0.5
shear_matrix = np.float32([[1, shear_factor, 0], [0, 1, 0]])

# Apply the shearing
sheared_image = cv2.warpAffine(image, shear_matrix, (int(image.shape[1] * 1.5), image.shape[0]))

# Display the original and sheared images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(sheared_image)
plt.title('Sheared Image')
plt.axis('off')

plt.show()
```
