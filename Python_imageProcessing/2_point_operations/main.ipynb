### histogram

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('nature.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(gray_image.shape)
print("no.of pixel: ", 893*720)

# Calculate histogram
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])
# [0,256)

# Normalize histogram for better visualization
histogram /= np.sum(histogram)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.plot(histogram, color='black')
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Normalized Frequency')
plt.xlim([0, 255])
plt.xticks(np.arange(0, 256, 10))
plt.yticks(np.arange(0, 0.2, 0.02))
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Thresholding

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('nature.jpg', cv2.IMREAD_GRAYSCALE)

# Define lower and upper threshold values
lower_threshold = 180
upper_threshold = 255

# Apply range thresholding
thresholded_image = cv2.inRange(image, lower_threshold, upper_threshold)

# Display original and thresholded images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(thresholded_image, cmap='gray')
plt.title(f'Thresholded Image: {lower_threshold}-{upper_threshold}')
plt.axis('off')

plt.tight_layout()
plt.show()
```

### Contrast Adjustment

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('nature.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Increase contrast by scaling pixel values
contrast_increase = 1.5

# linear transformation to an image, used to adjust the contrast and brightness of an image 
# by scaling and shifting pixel values {output_pixel = alpha * input_pixel + beta} range[0, 255]
contrast_image = cv2.convertScaleAbs(image, alpha=contrast_increase, beta=0)

# Display the original and contrast-adjusted images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(contrast_image)
plt.title('Contrast Adjusted Image')
plt.axis('off')

plt.show()
```

### contrast stretch

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('art.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
print(image.shape)

# Convert to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
print(893* 720)

# Perform contrast stretching
# min_intensity = np.min(gray_image)
# max_intensity = np.max(gray_image)
min_intensity =  220
max_intensity =255
print(min_intensity, max_intensity)

# fromula
stretched_image = 255 * ((gray_image - min_intensity) / (max_intensity - min_intensity))
# recliping

stretched_image = stretched_image.astype(np.uint8)

# Display original and stretched images
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.imshow(gray_image,cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(stretched_image, cmap='gray')
plt.title('Contrast Stretched Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Calculate histogram
histogram = cv2.calcHist([stretched_image], [0], None, [256], [0, 256])

# Normalize histogram for better visualization
histogram /= np.sum(histogram)

# Plot histogram
plt.figure(figsize=(10, 6))
plt.plot(histogram, color='black')
plt.title('Histogram of Pixel Intensities')
plt.xlabel('Pixel Intensity')
plt.ylabel('Normalized Frequency')
plt.xlim([0, 255])
plt.xticks(np.arange(0, 256, 10))
plt.yticks(np.arange(0, 0.2, 0.02))
plt.grid(True)
plt.tight_layout()
plt.show()
```

### Gamma Correction

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib
# print(image/255)

# Apply gamma correction
gamma = 0.5

#f(n)=x`0.5`

# non-linear operation used to encode and decode the luminance or tristimulus values in an image. 
# It is used to adjust the brightness of an image by altering the intensity of the pixels.
gamma_corrected = np.array(255 * (image / 255) ** gamma, dtype='uint8')

# Display the original and gamma-corrected images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(gamma_corrected)
plt.title('Gamma Corrected Image')
plt.axis('off')

plt.show()
```

### Brightness Adjustment

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load an image
image = cv2.imread('test_img.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert BGR to RGB for Matplotlib

# Increase brightness by adding a constant value to all pixels
brightness_increase = 50
bright_image = cv2.convertScaleAbs(image, alpha=1, beta=brightness_increase)

# Display the original and brightened images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(bright_image)
plt.title('Brightened Image')
plt.axis('off')

plt.show()
```

### templet matching

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the larger image and template image
image = cv2.imread('object.jpg', cv2.IMREAD_GRAYSCALE)
template = cv2.imread('art.jpg', cv2.IMREAD_GRAYSCALE)
w, h = template.shape[::-1]

# Apply template matching
result = cv2.matchTemplate(image, template, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

# Draw a rectangle around the matched region
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(image, top_left, bottom_right, 255, 2)

# Display the result
plt.imshow(image, cmap='gray')
plt.title('Template Matching Result')
plt.axis('off')
plt.show()
```
